```python
# í•™ìƒ ìˆ˜
num_students = 5

# í•™ìƒ ì •ë³´ ì €ì¥ ë¦¬ìŠ¤íŠ¸
students = []

# ì…ë ¥ ë°›ê¸°
for i in range(num_students):
    print(f"\n{i+1}ë²ˆì§¸ í•™ìƒ ì •ë³´ ì…ë ¥")
    name = input("ì´ë¦„: ")

    # Function to get valid integer input for a subject
    def get_score(subject):
        while True:
            try:
                score_str = input(f"{subject} ì ìˆ˜: ")
                score = int(score_str)
                return score
            except ValueError:
                print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤. ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")

    korean = get_score("êµ­ì–´")
    math = get_score("ìˆ˜í•™")
    science = get_score("ê³¼í•™")

    student = {
        "ì´ë¦„": name,
        "êµ­ì–´": korean,
        "ìˆ˜í•™": math,
        "ê³¼í•™": science,
        "í•©ê³„": korean + math + science,
        "í‰ê· ": round((korean + math + science) / 3, 2)
    }
    students.append(student)

# ê°œì¸ë³„ ì„±ì  ì¶œë ¥
print("\nğŸ§‘â€ğŸ“ í•™ìƒë³„ ì„±ì ")
print("-" * 40)
for s in students:
    print(f"ì´ë¦„: {s['ì´ë¦„']}, êµ­ì–´: {s['êµ­ì–´']}, ìˆ˜í•™: {s['ìˆ˜í•™']}, ê³¼í•™: {s['ê³¼í•™']}, í•©ê³„: {s['í•©ê³„']}, í‰ê· : {s['í‰ê· ']}")

# ê³¼ëª©ë³„ í•©ê³„ ë° í‰ê·  ê³„ì‚°
total_korean = sum(s['êµ­ì–´'] for s in students)
total_math = sum(s['ìˆ˜í•™'] for s in students)
total_science = sum(s['ê³¼í•™'] for s in students)

avg_korean = round(total_korean / num_students, 2)
avg_math = round(total_math / num_students, 2)
avg_science = round(total_science / num_students, 2)

# ê³¼ëª©ë³„ ì´ì  ë° í‰ê·  ì¶œë ¥
print("\nğŸ“Š ê³¼ëª©ë³„ ì´ì  ë° í‰ê· ")
print("-" * 40)
print(f"êµ­ì–´ - ì´ì : {total_korean}, í‰ê· : {avg_korean}")
print(f"ìˆ˜í•™ - ì´ì : {total_math}, í‰ê· : {avg_math}")
print(f"ê³¼í•™ - ì´ì : {total_science}, í‰ê· : {avg_science}")
```
